```python
from langgraph.graph import START, END, StateGraph
from typing import TypedDict
```

```python
#State = data passed between nodes

class MyState(TypedDict):
    message: str
```

```python
#Node 1

def greet(state: MyState) -> MyState:
    state["message"] = "Hello Jyoti!"
    return state
```

```python
# Node 2 
def ask_question(state: MyState) -> MyState:
    state["message"] += "How are you learning Langgraph?"
    return state
```

```python
# Build graph

graph = StateGraph(MyState)

graph.add_node("greet", greet)
graph.add_node("ask_question", ask_question)

graph.add_edge(START, "greet")
graph.add_edge("greet", "ask_question")
graph.add_edge("ask_question", END)

app = graph.compile()
```

```python
# Run the graph

output = app.invoke({"message": ""})
print(output)
```
