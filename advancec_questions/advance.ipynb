```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
```

```python
# ✅ State

class MyState(TypedDict):
    question: str
    answer: str
```

```python
ADVANCED_ANSWERS = {
    "what is ai": "AI (Artificial Intelligence) enables machines to perform tasks that normally require human intelligence.",
    "what is machine learning": "Machine Learning helps computers learn patterns from data automatically.",
    "what is deep learning": "Deep Learning uses neural networks with many layers to learn complex patterns.",
    "how langgraph works": "LangGraph works using Nodes (functions), Edges (connections), and State (data passed between nodes).",
    "what is a node": "A node is a simple Python function that performs one step in your workflow.",
    "what is an edge": "Edges decide the order of execution between nodes.",
    "what is state": "State is the data that flows through nodes in the graph.",
    "how this code works": "The code creates a graph with nodes and edges. Each node updates the state, and LangGraph decides the workflow order."
}
```

```python
def advanced_answer(state: MyState) -> MyState:
    q = state["question"].lower().strip()
    
    for key in ADVANCED_ANSWERS:
        if key in q:
            state["answer"] = ADVANCED_ANSWERS[key]
            return state

    # default response
    state["answer"] = "Sorry, I don't have an advanced answer for this yet."
    return state
```

```python
# ✅ Build graph
graph = StateGraph(MyState)
graph.add_node("advanced_answer", advanced_answer)
graph.add_edge(START, "advanced_answer")
graph.add_edge("advanced_answer", END)
app = graph.compile()
```

```python
result = app.invoke({"question": "What is AI?", "answer": ""})
print(result)
```

```python
result = app.invoke({"question": "What is state ?", "answer": ""})
print(result)
```

```python
result = app.invoke({"question": "What is surgery means?", "answer": ""})
print(result)
```
