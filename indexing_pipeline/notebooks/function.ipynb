```python
def is_prime(number:int) -> bool:
    """Checks if the number is prime or not

    Args:
        number (int): number

    Returns:
        bool: True if prime False otherwise
    """
    if number < 2:
        return False
    result = True
    index = 2
    while index <= number // 2:
        if number % index == 0:
            result = False
            break
        index += 1
    return result
```

```python
is_prime(7)
```

```python
import math
factorial_5 = math.factorial(5)
factors = [i for i in range(1, factorial_5+1) if factorial_5 % i == 0]
print(factors)
```

```python
def factorial2(number: int) -> int:
    """Calculates factorial of a number

    Args:
        number (int): number passed

    Returns:
        int: factorial
    """
    if number == 0:
        return 1
    return number * factorial(number-1)
```

```python
factorial2(0)
```

```python
factorial2(5)
```
